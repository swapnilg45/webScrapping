```python
import requests, pandas 
from bs4 import BeautifulSoup

r = requests.get("https://www.century21.com/real-estate/newark-ca/LCCANEWARK/")
c = r.content

soup = BeautifulSoup(c, "html.parser")

all = soup.find_all("div", {"class":"property-card-primary-info"})
#all
all[0].find("a", {"class":"listing-price"}).text.replace('\n', "").strip()

```

```python
## extracting the prices
l = []
for items in all:
    d = {}
    try:
        d["Price"] = items.find("a", {"class":"listing-price"}).text.replace('\n', "").strip()
    except:
         d["Price"] = None
        
    try:
        d["Address"] = items.find("div", {"class":"property-address"}).text.replace('\n', "").strip()
    except:
        d["Address"] = None
        
    try:
        d["City"] = items.find("div", {"class":"property-city"}).text.replace('\n', "").strip()
    except:
        d["City"] = None
        
    try:
        d["Square_Feet"] = (items.find("div", {"class":"property-sqft"}).text.replace('\n', "").strip())
    except:
        d["Square_Feet"] = None
        
    try:
        d["No_of_beds"] = items.find("div", {"class":"property-beds"}).text.replace('\n', "").strip()[0]
    except:
        d["No_of_beds"] = None
        
    try:
        d["No_of_baths"] = items.find("div", {"class":"property-baths"}).text.replace('\n', "").strip()[0]
    except:
        d["No_of_baths"] = None
        
    l.append(d)
        
    print(" ")
```

```python
df = pandas.DataFrame(l)
df
```
